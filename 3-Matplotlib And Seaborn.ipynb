## Matplotlib

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

##### scatterplot :

```python
plt.scatter(a,b,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
#plt.savefig('Test.png')
```

##### plt plot

```python
a=np.arange(5,10)
b=a*a
```

```python
#plt.plot(a,b,'b*--')

plt.plot(a,b,'b*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('2d graph')
```

##### plt subplots :

```python
a=np.arange(10,20)
b=a*2
```

```python
p=np.arange(1,10)
q=p*p
```

```python
#plt.subplot(nrow,ncol,pos)

plt.subplot(2,2,1)
plt.plot(a,b,'go')

plt.subplot(2,2,2)
plt.plot(p,q,'r*--')

plt.subplot(2,2,3)
plt.plot(a,b,'b*--')

plt.subplot(2,2,4)
plt.plot(p,q,'yo')
```

##### Other plots :

```python
x= np.arange(0,4*np.pi,0.1)
y= np.sin(x)

plt.plot(x,y)
plt.title('Sine wave')
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)

plt.subplot(2,1,1)
plt.plot(x,y,c='g',linestyle="dashed")
plt.title("sine wave")


a= np.arange(0,4*np.pi,0.1)
b= np.cos(a)
plt.subplot(2,1,2)
plt.plot(a,b,'r')
plt.title('cosine wave')
```

##### Barplot :

```python
x1 = [2,6,10] 
y1 = [11,16,9]  

x2 = [3,7,11] 
y2 = [6,15,7] 

plt.bar(x1,y1,color='g')
plt.bar(x2,y2,color='b')
plt.title("Bar graph")
plt.xlabel('x-axis')
plt.ylabel('y-axis')

```

##### Histogram :

```python
a=np.array([6,80,99,11,22,45,76,12,47,55])
plt.hist(a)
plt.title("histogram")
```

##### Boxplot :

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

##### Pie-chart :

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [220, 110, 240, 120]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

explode = (0.4, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```

## Seaborn

```python
import seaborn as sns
```

### Distribution plots

i-distplot(univariate analysis)

ii-jointplot(bivariate analysis)

iii-pairplot(multivariate analysis)

```python
df= pd.read_csv("tips.csv")
#df=sns.load_dataset("tips")
```

```python
df.head(5)
```

=>Univariate analysis(Analysis for one feature)

=>Bivariate analysis(Analysis for more than one feature i.e. for two features)

=>multivariate analysis(Analysis of features with respect to other features)

##### Correlation with heatmap :

A correlation helps us to find the inter-relation between all the features having continous variables.

It is very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

##### Joint plot :

A join plot allows to study the relationship between 2 numeric variables.

```python
sns.jointplot(x=df["tip"],y=df["total_bill"],kind="hex")

#sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
```

##### Pair plot :

A “pairs plot” is also known as a scatterplot.

In this, all variables are paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue="day")
```

##### Distplot :

```python
sns.distplot(df["total_bill"])
```

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

### Categorical plots

i-boxplot

ii-violinplot

iii-countplot

iv-bar plot

##### Count plot :

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(df['day'])
```

```python
sns.countplot(y=df['day'])
```

##### Bar plot :

```python
sns.barplot(x=df['total_bill'],y=df['day'])
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(y='total_bill',x='smoker',data=df)
```

##### Box plot :

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df)
```

```python
sns.boxplot(df['sex'],df['total_bill'],palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(df['total_bill'],df['day'],hue=df['smoker'])
```

```python
sns.boxplot('total_bill','day',data=df,hue='sex')
```

##### Violin plot :

It helps us to see both the distribution of data in terms of Kernel density estimation and the box plot.

```python
sns.violinplot('total_bill','day',data=df,palette='rainbow')
```

```python
sns.violinplot(y=df['total_bill'],x=df['sex'])
```
